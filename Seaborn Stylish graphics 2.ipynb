# Mission 1

Reproduce the graphic below using Seaborn

```
import numpy as np
import matplotlib.pyplot as plt


rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,cmap='viridis')
```

# Mission 2

Reproduce the graphic below using Seaborn

```
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(19680801)

data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots()
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set(xlabel='entry a', ylabel='entry b')
plt.show()
```

# Mission 4

Reproduce the graphic below using Seaborn

```
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

np.random.seed(19680801)


all_data = [np.random.normal(0, std, 100) for std in range(6, 10)]

axes[0].violinplot(all_data,
                   showmeans=False,
                   showmedians=True)
axes[0].set_title('Violin plot')

axes[1].boxplot(all_data)
axes[1].set_title('Box plot')

for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y + 1 for y in range(len(all_data))])
    ax.set_xlabel('Four separate samples')
    ax.set_ylabel('Observed values')

plt.setp(axes, xticks=[y + 1 for y in range(len(all_data))],
         xticklabels=['x1', 'x2', 'x3', 'x4'])
plt.show()
```

```
sns.set_style('white')
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

np.random.seed(19680801)

all_data = [np.random.normal(0, std, 100) for std in range(6, 10)]
xticks=['x1', 'x2', 'x3', 'x4']
sns.violinplot(data=all_data, ax=axes[0], color="skyblue") 
axes[0].set_title('Violin plot')
axes[0].set_xticklabels(xticks)


id = sns.boxplot(xticks, all_data,ax=axes[1], width=0.5, showmeans=False, medianprops={'color':'orange'} )
axes[1].set_title('Box plot')
for box in id.artists:
    box.set_facecolor("white")

for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xlabel('Four separate samples')
    ax.set_ylabel('Observed values')

plt.show()
```

# Mission 6

What does the method **sns.pairplot()** do?

This function will create a grid of Axes such that each numeric variable in data will by shared in the y-axis across a single row and in the x-axis across a single column. The diagonal Axes are treated differently, drawing a plot to show the univariate distribution of the data for the variable in that column.
