# Mission 1

Create the infinity symbol using Seaborn. You can use Cosine or any other function. Don't worry about making a perfect curve, just make sure the two lines intersect.

![symbole de l'infini](https://upload.wikimedia.org/wikipedia/commons/thumb/c/c8/Infinite.svg/1200px-Infinite.svg.png)

```
import seaborn as sns
from math import pi
import numpy as np
import matplotlib.pyplot as plt

sns.set()
Data_infinity = np.linspace(-pi, pi, 1000)
sns.lineplot(Data_infinity, np.sin(Data_infinity))
sns.lineplot(Data_infinity, np.cos(Data_infinity+np.pi/2))
plt.gcf().set_size_inches(10,5)
sns.despine(top = False, bottom = False, left = False, right = False)
```

# Mission 3

Recreate the visual below using Seaborn

```
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels)) 
width = 0.35  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```

```
sns.set()
Data_score_by_group_and_gender = {'Groups':['G1', 'G1', 'G2', 'G2', 'G3','G3', 'G4', 'G4', 'G5', 'G5'],                   
        'Gender':['Men','Women','Men','Women','Men','Women','Men','Women','Men','Women'],
        'Scores': [20, 25, 34, 32, 30 ,34, 35, 20, 27, 25]}
sns.barplot(x='Groups', y='Scores', data = Data_score_by_group_and_gender, hue='Gender', palette="bright")
```

# Mission 4

Make detailed comments about the below code

```
import pandas as pd # Importer les bibliothèques Pandas et Seaborn
import seaborn as sns
sns.set() # Activer Seaborn

df = sns.load_dataset("brain_networks", header=[0, 1, 2], index_col=0) # Définir le Dataframe avec Seaborn

used_networks = [1, 5, 6, 7, 8, 12, 13, 17] # Création d'une liste nommée "used_networks"
used_columns = (df.columns.get_level_values("network") # Sélectionner les critères
                          .astype(int)
                          .isin(used_networks))
df = df.loc[:, used_columns] # Sélectionner dans df certains critères

network_pal = sns.husl_palette(8, s=.45) # Choisir les couleurs
network_lut = dict(zip(map(str, used_networks), network_pal)) # Création d'un Dictionnaire avec "used_networks" et avec les couleurs "networl_pal"

networks = df.columns.get_level_values("network") # Appliquer "network_lut" à "networl_colors"
network_colors = pd.Series(networks, index=df.columns).map(network_lut) # Appliquer "network_lut" à "networl_colors"

sns.clustermap(df.corr(), center=0, cmap="vlag", # Choix du type de graphique, avec couleurs, taille, titres, centrage etc...
               row_colors=network_colors, col_colors=network_colors,
               linewidths=.75, figsize=(13, 13))
```
