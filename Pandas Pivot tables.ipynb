# Execute the code below

```
import pandas as pd
import numpy as np
link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/weather2019.csv"
df_weather = pd.read_csv(link)
```

# You have now a DataFrame with lots of weather data

```
df_weather.head()
```

# PIVOT TABLE

## It's like in Excel, but with Pandas

There are 2 syntaxes, choose your preferred :
- As a method : YOUR_DATAFRAME.pivot_table()
- As a function : pd.crosstab()

```
# Example
df_weather.pivot_table(values = 'MAX_TEMPERATURE_C', index = 'MONTH', aggfunc = 'mean')
```

## Create a Pivot table with :
- Months on rows
- the maximum of Max Temperature on values
- Opinion on columns

```
# Hop !
df_weather.pivot_table(values = 'MAX_TEMPERATURE_C', index = 'MONTH', columns = 'OPINION')
```

## Pivot table with :
- Days on rows
- Precipitation on values
- Months on columns


```
# Hop !
df_weather.pivot_table(values = 'PRECIP_TOTAL_DAY_MM', index = 'DAY', columns = 'MONTH')
```

## Pivot table with :
- Months on rows
- the sum of Precipitations on values

**Then display it as a column chart.**

```
# Hop !
df_precipitations = df_weather.pivot_table(values = 'PRECIP_TOTAL_DAY_MM', index = 'MONTH', aggfunc = "sum")
df_precipitations
```

```
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(palette = 'dark')

Column_Chart_Precipitations = sns.barplot(x = df_precipitations.index, y = 'PRECIP_TOTAL_DAY_MM', data = df_precipitations)

plt.show()
```

## Pivot table with :
- Months on rows
- the average of Max Temperature on a first column of values
- the average of Sunhours on a second column of values

```
# Hop !

df_MaxTem_Sunhour = df_weather.pivot_table(values = ['MAX_TEMPERATURE_C', 'SUNHOUR'] , index = 'MONTH')

df_MaxTem_Sunhour
```

# Chart
Draw a chart from the previous pivot table with :
- Temperature in a blue column chart
- Sunhours in a red line plot

```
# Your code here
plt.bar(df_MaxTem_Sunhour.index, 
        df_MaxTem_Sunhour['MAX_TEMPERATURE_C'], 
        color= 'green', 
        label = 'Max Temperature')

plt.plot(df_MaxTem_Sunhour.index, 
         'SUNHOUR', 
         data = df_MaxTem_Sunhour, 
         color = 'red', 
         linewidth = 3, 
         label = 'Sunhour')

plt.legend()
plt.show()
```
